# **ENCODING THE CATEGORICAL COLUMNS :**
Encoding categorical columns is essential in machine learning as it transforms non-numeric data into a format that algorithms can understand, enabling the utilization of categorical information for modeling and improving prediction accuracy.

Machine Learning algorithms can only understant NUMERICAL VALUES, not CATEGORICAL VALUES. (CATEGORICAL VALUES : String type Values)

**STORE FEATURE MATRIX (INPUT COLUMNS) & RESPONSE(TARGET)(ACTUAL OUTPUT) :**

Storing the feature matrix (input columns) and response (actual output) separately is crucial for training and evaluating machine learning models, as it allows for clear separation of input data and target values, enabling effective model training and assessment of predictive accuracy.

```python
print(dataset['CarName'].unique())
```

```python
input = dataset.drop(['CarName','price'], axis=1)
output = dataset['price']

# print(input)
print(output)
```



---



**SPLITTING DATASET INTO TRAINING & TESTING SETS :**

Splitting the dataset into training and testing sets is vital for assessing a machine learning model's performance. It helps gauge how well the model generalizes to new, unseen data, preventing overfitting and ensuring reliable predictions.

```python
from sklearn.model_selection import train_test_split
inputTRAIN, inputTEST, outputTRAIN, outputTEST = train_test_split(input, output, test_size=0.2)
```

# **MODEL**

```python
dataset.head()
```

As TARGET COLUMN CONTAIN A CONTINUES VALUES :=> THIS IS A REGRESSION PROBLEM .

**IMPORTING REGRESSION MODELS : **

1. LiearRegression
2. RandomForestRegressor
3. GradientBoostingRegressor
4. XGBRegressor

IMPORTING ALGORITHMS FROM LIBRARIES :

```python
 from sklearn.linear_model import LinearRegression
 from sklearn.ensemble import RandomForestRegressor
 from sklearn.ensemble import GradientBoostingRegressor
 from xgboost import XGBRegressor
```

**TRAINNING TO THE MODELS :**

```python
# TRAINNING LinearRegression MODEL :
LR_Model = LinearRegression()
LR_Model.fit(inputTRAIN, outputTRAIN)

# TRAINNING RandomForestRegressor MODEL :
RFR_Model = RandomForestRegressor()
RFR_Model.fit(inputTRAIN, outputTRAIN)

# TRAINNING GradientBoostingRegressor MODEL :
GBR_Model = GradientBoostingRegressor()
GBR_Model.fit(inputTRAIN, outputTRAIN)

# TRAINNING XGBRegressor MODEL :
XGBR_Model = XGBRegressor()
XGBR_Model.fit(inputTRAIN, outputTRAIN)
```

**PREDICTING OUTPUT OF ALL MODELS :**

```python
LR_Prediction = LR_Model.predict(inputTEST)
RFR_Prediction = RFR_Model.predict(inputTEST)
GBR_Prediction = GBR_Model.predict(inputTEST)
XGBR_Prediction = XGBR_Model.predict(inputTEST)
```

**ACCURACY SCORES OF EACH MODEL :**

```python
from sklearn import metrics
```

1) LinearRegression :

```python
LR_Score = metrics.r2_score(outputTEST,LR_Prediction)*100
print("LinearRegression ACCURECY : ",LR_Score)
```

2) RandomForestRegressor :

```python
RFR_Score = metrics.r2_score(outputTEST,RFR_Prediction)*100
print("RandomForestRegressor ACCURECY : ",RFR_Score)
```

3) GradientBoostingRegressor :

```python
GBR_Score = metrics.r2_score(outputTEST,GBR_Prediction)*100
print("GradientBoostingRegressor ACCURECY : ",GBR_Score)
```

4) XGBRegressor :

```python
XGBR_Score = metrics.r2_score(outputTEST,XGBR_Prediction)*100
print("XGBRegressor ACCURECY : ",XGBR_Score)
```
